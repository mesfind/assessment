### All workshops in 2017

```python
workshops2017 = workshops['year'] == 2017
workshops2017 = workshops[workshops2017]
countries2017 = workshops2017['slug'].groupby([workshops2017['country']])
countries2017 = countries2017.count()


countries2017.plot.bar()

```

```python
workshops2018 = workshops['year'] == 2018
workshops2018 = workshops[workshops2018]

workshops2018Q1 = workshops2018['month'] <= 3
workshops2018Q1 = workshops2018[workshops2018Q1]


countries2018Q1 = workshops2018Q1['slug'].groupby([workshops2018Q1['country']])


countries2018Q1 = workshops2018Q1['slug'].groupby([workshops2018Q1['country']])
countries2018Q1 = countries2018Q1.count()
countries2018Q1
countries2018Q1.plot.bar()
```

```python
workshops2017Q1 = workshops2017['month'] <= 3
workshops2017Q1 = workshops2017[workshops2017Q1]


countries2017Q1 = workshops2017Q1['slug'].groupby([workshops2017Q1['country']])


countries2017Q1 = workshops2017Q1['slug'].groupby([workshops2017Q1['country']])
countries2017Q1 = countries2017Q1.count()
countries2017Q1
countries2017Q1.plot.bar()
```

```python
fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

countries2017Q1.plot(kind='bar', color='red', ax=ax, width=width, position=1)
countries2018Q1.plot(kind='bar', color='blue', ax=ax2, width=width, position=0)

ax.set_ylabel('2017Q1')
ax2.set_ylabel('2018Q1')

plt.show()





```
